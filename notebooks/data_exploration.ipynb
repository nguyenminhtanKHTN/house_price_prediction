## Data exploration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/raw/housing.csv')
display(df.head())
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Phân phối giá nhà
plt.figure(figsize=(8, 6))
sns.histplot(df['median_house_value'], bins=30)
plt.title('Phân phối giá nhà')
plt.savefig('../results/feature_distribution.png')
plt.show()
```

```python
numeric_features = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_features = df.select_dtypes(include=[object]).columns.tolist()

# Ma trận tương quan
plt.figure(figsize=(10, 8))
sns.heatmap(df[numeric_features].corr(), annot=True, cmap='coolwarm')
plt.title('Ma trận tương quan')
plt.savefig('../results/correlation_matrix.png')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df['median_house_value'])
plt.title('Boxplot of House Prices (MedHouseVal)')
plt.savefig('../results/boxplot_medhouseval.png')
plt.show()
```
